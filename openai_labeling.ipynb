## CODE 8K

````python
CODE8K_PROMPT = """Solve the following math problem algorithmically. The answer should be an integer. 
```
{problem}
```
First, briefly think through how to approach the problem algorithmically.
Then write a single executable Python program with fenced code block (```python and ```). You may use sympy.
The last line of your code solution should be `print(answer)`. No need for further explanation.
"""

prompts = [CODE8K_PROMPT.format(problem=prob) for prob in df["problem"]]
answers = df["answer"].tolist()
answers = [answer%1000 for answer in answers]
````

````python
res = generate_batch(prompts, batch_size=4, estimate=False, save_intermediate=True)
````

````python
preds = [extract_answer(res[i]["completion"]) for i in range(len(res))]
print(preds[:10])
print(answers[:10])
````

````python
df["completions"] = [res[i]["completion"] for i in range(len(res))]
df["pred"] = preds
df
````

````python
df["correct"] = df.apply(lambda x: x["answer"]%1000 == x["pred"]%1000, axis=1)
df
````

````python
df["o3mini_code_solution"] = df.apply(lambda x: x["completions"] if x["correct"] else "", axis=1)
df
````

````python
df = df[["problem", "answer", "claude_solution","o3mini_code_solution"]]
df = df.rename(columns={"claude_solution": "claude_en_solution"})
df
````
